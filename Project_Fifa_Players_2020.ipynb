# Introduction

In this project, we will explore and analyze FIFA Players 2020 dataset. The dataset shows the number of football players that actively playing football as 2020 in respect of their Club teams and National Country. Our emphasis will be cleaning and analysing dataset, predictions, visualization and provide conclusions.

### 5. Analyze the FIFA Dataset to predict the World’s Best Playing Eleven in 2020!!

World’s Best Playing Eleven of the Dataset:
- The Best Attacker
- The Best Mid-Fielders
- The Best Defenders
- The Best Goalkeeper


```python
# Defining the players by their role in football
players_position = ('pace', 'shot', 'passing', 'drib', 'dfend', 'physic','gkdiv', 'gkhan', 'gkkik', 
                    'gkflx', 'gkspd', 'gkpst', 'crossing', 'finishing', 'heading', 'control', 
                    'agility', 'reactions', 'jumping', 'stamina', 'aggression', 'interceptions', 
                    'position', 'vision', 'penalties', 'composure', 'marking', 'tackle')
# consider the best 8 features of every position of the dataset
for i, val in players.groupby(players['tpst'])[players_position].mean().iterrows():
    print('tpst {}: {}, {}, {}, {}, {}, {}, {}, {}'.format(i, * tuple(val.nlargest(8).index)))
```

- Considering a playing formation of 4-3-3 which implies to be a goalkeeper, 4 defenders, 3 midefielders and 3 attackers.
- 1 goalkeeper, 2 center back, 1 left back, 1 right back, 1 center midfielder, 1 left midfielder, 1 right midfielder, 1 left winger, 1 right winger and 1 stricker

```python
# finding the goalkeeper for the team (team_gk)
players['team_gk'] = (players.gkflx + players.gkdiv + players.gkpst + players.gkhan + 
                      players.gkkik + players.reactions +players.jumping + players.composure)

plt.figure(figsize=(15,6))
lads = players[(players['tpst']=='GK')].sort_values('team_gk', ascending=False)[:3]
x = np.array(list(lads['name']))
y = np.array(list(lads['team_gk']))
sns.barplot(x, y, palette="colorblind")
plt.xlabel('Goalkeeper Names')
plt.ylabel('Goalkeeper Attribute')
```

```python
# finding the best Right Center Back defender for the team (team_RCB)
players['team_RCB'] = (players.physic + players.aggression + players.tackle + players.heading + 
                       players.dfend + players.marking + players.jumping + players.interceptions)
plt.figure(figsize=(15,6))
lads = players[(players['tpst']=='RCB') | (players['tpst']=='RB')].sort_values('team_RCB', ascending=False)[:3]
x = np.array(list(lads['name']))
y = np.array(list(lads['team_RCB']))
sns.barplot(x, y, palette="colorblind")
plt.xlabel('Right Center Back Names')
plt.ylabel('Right Center Back Attributes')
```
